# Adagrad: Adaptive Gradient Algorithm

**Adagrad** (Adaptive Gradient Algorithm) is an optimization method designed to dynamically adjust the learning rate for each parameter in a machine learning model based on the magnitude of the gradients encountered during training.

---

## Theoretical Breakdown of Adagrad

### Problem with Traditional Gradient Descent
In traditional gradient descent, a single fixed learning rate is applied uniformly across all parameters, leading to several issues:

- **Frequent Updates**: 
  - Parameters that are frequently updated may overshoot the optimal point if the learning rate is too high.

- **Infrequent Updates**: 
  - Parameters that receive fewer updates might require a higher learning rate to make meaningful progress.

Adagrad addresses these challenges by adapting the learning rate for each parameter based on historical gradients.

---

## Key Features of Adagrad

1. **Adaptive Learning Rate**  
   - Adagrad modifies the learning rate for each parameter individually, scaling it based on the accumulated squared gradients over time. This means:
     - **Frequent Updates**: Parameters with frequent updates receive smaller adjustments.
     - **Infrequent Updates**: Parameters with fewer updates receive larger adjustments, helping balance learning rates across parameters.

2. **Automatic Adjustment**  
   - Adagrad eliminates the need for manual tuning of learning rates for different parameters by automatically adjusting learning rates based on observed gradients, simplifying implementation.

3. **Enhanced Stability in Training**  
   - By reducing the learning rate for parameters with large accumulated gradients, Adagrad helps prevent overshooting, promoting stable convergence.

---

## Adaptive Learning Rate in Adagrad

Adagrad adapts the learning rate based on the accumulated squared gradients for each parameter, adjusting the learning rate for each parameter individually:

- **Increases step size** for parameters with smaller accumulated gradients.
- **Decreases step size** for parameters with larger accumulated gradients.

However, the learning rate tends to decrease **monotonically** over time, which can make it challenging for the algorithm to continue learning effectively after many iterations.

---

## Limitations of Adagrad

- **Aggressive Learning Rate Decay**  
  - Accumulated gradients grow over time, leading to increasingly smaller effective learning rates.
  - This aggressive decay can cause the algorithm to stop learning prematurely, particularly in tasks requiring sustained updates.

---


### Adagrad in Practice

- **Applicability**: Due to its limitations regarding aggressive learning rate decay, Adagrad may not be suitable for all applications, particularly those that require a consistent learning rate over time.

- **Variants**: Algorithms like **RMSProp** and **Adam** have been developed to address these limitations by modifying how accumulated gradients are handled, resulting in a more stable learning rate throughout training.

### 1. NumPy: Used for Array Manipulation, Square Root Calculations, and Random Number Generation
The code utilizes the NumPy library for efficient array operations, mathematical calculations (like square roots), and generating random numbers.

### 2. Define the Objective Function
The function `objective(x, y)` computes the value of the objective function, which is defined as:

$$
f(x, y) = x^2 + y^2
$$

This function represents a simple quadratic surface, where the minimum occurs at the point \((0, 0)\).

### 3. Define the Derivative of the Objective Function
The function `derivative(x, y)` calculates the gradient of the objective function 

$$
\nabla f(x, y) = \begin{bmatrix} \frac{\partial f}{\partial x} \\ \frac{\partial f}{\partial y} \end{bmatrix} = \begin{bmatrix} 2x \\ 2y \end{bmatrix}
$$

The gradient points in the direction of the steepest ascent, thus moving in the opposite direction leads to the minimum of the function.

### 4. Gradient Descent Algorithm with Adagrad
The function `adagrad(objective, derivative, bounds, n_iter, step_size)` implements the Adagrad optimization algorithm.

#### 4.1 Generate Initial Solution
This line generates a random starting point within specified bounds. The `bounds` variable is an array, for instance, \([[-1.0, 1.0], [-1.0, 1.0]]\), indicating the allowed range for \(x\) and \(y\).

#### 4.2 Initialize Tracking Variables
- **sq_grad_sums**: A list to track the sum of squared gradients for each parameter, initialized to zero. This variable is crucial for the Adagrad algorithm as it helps adjust the learning rate dynamically.

### 5. Gradient Descent Loop
The loop runs for a specified number of iterations (`n_iter`).

#### 5.1 Calculate Gradient
Within each iteration, the gradient of the objective function at the current point `solution` is computed:

$$
\nabla f(x, y) = \begin{bmatrix} \frac{\partial f}{\partial x} \\ \frac{\partial f}{\partial y} \end{bmatrix} = \begin{bmatrix} 2x \\ 2y \end{bmatrix}
$$

#### 5.2 Update the Sum of Squared Gradients
For each variable, the squared gradient is accumulated:

$$
\text{sq\_grad\_sums}[i] += (\text{gradient}[i])^2
$$

This running total represents the accumulated information about the gradients, which is essential for adjusting the step size.

#### 5.3 Build New Solution
A new solution is computed one variable at a time. For each parameter:

1. **Calculate Step Size**:
   The step size alpha for each parameter is calculated using the accumulated squared gradients:

   $$
   \alpha = \frac{\text{step\_size}}{1 \times 10^{-8} + \sqrt{\text{sq\_grad\_sums}[i]}}
   $$

   This formula ensures that larger gradients lead to smaller updates, providing a mechanism for adaptive scaling of the step size.

2. **Update Variable**:
   The new value for each variable is computed as follows:

   $$
   \text{value} = \text{solution}[i] - \alpha \times \text{gradient}[i]
   $$

   The updated variable is then stored in `new_solution`.

#### 5.4 Update Solution
The new solution is converted into a NumPy array and replaces the previous solution:

$$
\text{solution} = \text{new\_solution}
$$

#### 5.5 Evaluate the Candidate Point
The objective function is evaluated at the new position:

$$
\text{solution\_eval} = f(\text{solution}[0], \text{solution}[1])
$$


```python
import numpy as np
from math import sqrt
from numpy import asarray
from numpy.random import rand, seed

# Objective function
def objective(x, y):
    return x**2.0 + y**2.0

# Derivative of the objective function
def derivative(x, y):
    return asarray([x * 2.0, y * 2.0])

# Gradient descent algorithm with Adagrad
def adagrad(objective, derivative, bounds, n_iter, step_size):
    # Generate an initial point randomly within bounds
    solution = bounds[:, 0] + rand(len(bounds)) * (bounds[:, 1] - bounds[:, 0])
    
    # List of the sum square gradients for each variable
    sq_grad_sums = [0.0 for _ in range(bounds.shape[0])]
    
    # Run the gradient descent
    for it in range(n_iter):
        # Calculate gradient
        gradient = derivative(solution[0], solution[1])
        
        # Update the sum of squared gradients
        for i in range(gradient.shape[0]):
            sq_grad_sums[i] += gradient[i]**2.0
            
        # Build a solution one variable at a time
        new_solution = []
        for i in range(solution.shape[0]):
            # Calculate the step size for this variable
            alpha = step_size / (1e-8 + sqrt(sq_grad_sums[i]))
            # Calculate the new position in this variable
            value = solution[i] - alpha * gradient[i]
            # Store this variable
            new_solution.append(value)
        
        # Update the solution
        solution = asarray(new_solution)
        
        # Evaluate the candidate point
        solution_eval = objective(solution[0], solution[1])
        
        # Report progress (as seen in the attached image)
        print(f'>{it+1} f({solution}) = {solution_eval:.5f}')
        
    return solution, solution_eval

# Seed the pseudo-random number generator
seed(1)

# Define the range for input
bounds = asarray([[-1.0, 1.0], [-1.0, 1.0]])

# Define the total iterations
n_iter = 50

# Define the step size
step_size = 0.1

# Perform the gradient descent search with Adagrad
best, score = adagrad(objective, derivative, bounds, n_iter, step_size)

# Print the final result
print('Done!')
print(f'f({best}) = {score:.6f}')
```

# Algorithm Steps

1. **Track Solutions**: Initializes an empty list called `solutions` to store all solutions found during the iterations.
  
2. **Initial Solution**: Generates a random starting point within the defined bounds using NumPy's `rand` function, which is scaled to fit the specified range for each variable.

3. **Squared Gradient Averages**: Initializes a list `sq_grad_sums` to hold the accumulated squared gradients for each variable, starting from zero.

4. **Iteration Loop**: The loop runs for `n_iter` iterations, updating the solution based on the calculated gradients.

5. **Gradient Calculation**: Computes the gradient at the current solution point using the derivative function:

   $$
\nabla f(x, y) = \begin{bmatrix} \frac{\partial f}{\partial x} \\ \frac{\partial f}{\partial y} \end{bmatrix} = \begin{bmatrix} 2x \\ 2y \end{bmatrix}
$$

# Gradient Update Loop

In each iteration, the algorithm performs the following steps:

- **Squared Gradient Update**: For each variable, the squared gradient is accumulated:

   $$
   \text{sq\_grad\_sums}[i] += g_i^2
   $$

   This represents the sum of the squared gradients for each variable, which is essential for adjusting the learning rate.

- **New Solution List**: Initializes an empty list `new_solution` to hold the updated values for each variable in the solution.

- **Variable Update Loop**: This inner loop iterates through each variable in the solution.

- **Step Size Calculation**: The adaptive step size alpha  for each variable is calculated using the formula:

   $$
   \alpha = \frac{\text{step\_size}}{1 \times 10^{-8} + \sqrt{\text{sq\_grad\_sums}[i]}}
   $$

   This formula ensures that larger gradients will result in smaller updates, allowing for adaptive scaling of the learning rate.

- **Change Calculation**: The change for the current variable is determined by multiplying the computed step size by the gradient:

   $$
   \Delta x_i = -\alpha \cdot g_i
   $$

- **Updating the Solution**: The new value for the current variable is computed by applying the change:

   $$
   \text{value} = \text{solution}[i] + \Delta x_i
   $$

- **Storing New Variable Value**: The computed new value is appended to the `new_solution` list.

- **Updating Current Solution**: After processing all variables, the new solution is converted to a NumPy array and assigned back to `solution`:

   $$
   \text{solution} = \text{new\_solution}
   $$

- **Candidate Evaluation**: The objective function is evaluated at the new solution:

   $$
   \text{solution\_eval} = f(\text{solution}[0], \text{solution}[1])
   $$

- **Progress Output**: The current iteration number, the solution vector, and the evaluation of the objective function at that solution are printed to track the optimization process.

- **Final Output**: After completing all iterations, the function returns the list of solutions found during the optimization process.


```python
# Gradient descent algorithm with Adagrad
def adagrad(objective, derivative, bounds, n_iter, step_size):
    # Track all solutions
    solutions = list()
    
    # Generate an initial point
    solution = bounds[:, 0] + rand(len(bounds)) * (bounds[:, 1] - bounds[:, 0])
    
    # List of the sum of square gradients for each variable
    sq_grad_sums = [0.0 for _ in range(bounds.shape[0])]
    
    # Run the gradient descent
    for it in range(n_iter):
        # Calculate gradient
        gradient = derivative(solution[0], solution[1])
        
        # Update the sum of the squared partial derivatives
        for i in range(gradient.shape[0]):
            sq_grad_sums[i] += gradient[i]**2.0
        
        # Build the new solution
        new_solution = list()
        for i in range(solution.shape[0]):
            # Calculate the learning rate for this variable
            alpha = step_size / (1e-8 + sqrt(sq_grad_sums[i]))
            # Calculate the new position in this variable
            value = solution[i] - alpha * gradient[i]
            new_solution.append(value)
        
        # Store the new solution
        solution = asarray(new_solution)
        solutions.append(solution)
        
        # Evaluate candidate point
        solution_eval = objective(solution[0], solution[1])
        
        # Report progress
        print('>%d f(%s) = %.5f' % (it, solution, solution_eval))
    
    return solutions
```

This loop iterates through each dimension of the gradient.

- **Moving Average Update**: This updates the moving average of the squared gradients using the formula:

$$
E[g_i^2] = \rho \cdot E[g_i^2] + (1 - \rho) \cdot g_i^2
$$

Here, $\rho$ is a decay factor that determines how much weight is given to the previous average versus the current square


- **New Solution List**: This initializes an empty list to hold the updated values for the solution variables.

- **Variable Update Loop**: This loop iterates through each variable in the solution.

- **Step Size Calculation**: The step size \( \alpha \) is computed using the AdaDelta formula:

$$
\alpha = \frac{\epsilon + E[\Delta x_i^2]}{\epsilon + E[g_i^2]}
$$

This step size is adaptive and depends on the moving averages of the squared parameter updates and squared gradients, ensuring that the updates adjust based on past behavior.

- **Change Calculation**: The change for the current variable is determined by multiplying the computed step size by the corresponding gradient

$$
E[\Delta x_i^2] = \rho \cdot E[\Delta x_i^2] + (1 - \rho) \cdot (\Delta x_i^2)

### Objective Function
This function defines the objective function 
$$
f(x, y) = x^2 + y^2
$$
This is a simple quadratic function representing a paraboloid. The goal is to minimize this function.

#### Mathematics Involved:
- The function is symmetric around the origin \((0, 0)\).
- The minimum value is \(0\), occurring at the point \((0, 0)\).

### Derivative of the Objective Function
This function calculates the gradient (first derivatives) of the objective function:
- The gradient 
$$
\nabla f(x, y) = \begin{bmatrix} \frac{\partial f}{\partial x} \\ \frac{\partial f}{\partial y} \end{bmatrix} = \begin{bmatrix} 2x \\ 2y \end{bmatrix}
$$

This vector points in the direction of the steepest ascent of the function.

### Adadelta Algorithm
**Parameters:**
- `objective`: The function we want to minimize.
- `derivative`: The gradient of the objective function.
- `bounds`: The lower and upper limits for the input variables (defined later).
- `n_iter`: Number of iterations to perform.
- `rho`: The decay factor for the moving averages.
- `ep`: A small constant to avoid division by zero 

**Initialization:**
- `solutions`: List to store the solutions found in each iteration.
- `solution`: Generates a random starting point within the defined bounds.
- `sq_grad_avg`: List to store the running average of squared gradients.
- `sq_para_avg`: List to store the running average of squared parameter updates.

### Main Loop of Adadelta
1. **Gradient Calculation**: The gradient at the current solution is calculated using the derivative function.
2. **Squared Gradient Average Update**:
   - For each component of the gradient, the squared value \(sg\) is computed.
   - The running average of squared gradients is updated using an exponential decay defined by rho:
   $$
   sq\_grad\_avg[i] = \rho \cdot sq\_grad\_avg[i] + (1 - \rho) \cdot sg
   $$


### Update Solution
1. **Step Size Calculation**:
   The step size \( \alpha \) is computed using the averages of squared parameters and gradients:
   $$
   \alpha = \frac{\epsilon + sq\_para\_avg[i]}{\epsilon + sq\_grad\_avg[i]}
   $$

2. **Parameter Update**:
   - The change is computed by scaling the gradient by the step size.
   - The squared parameter average is updated:
   $$
   sq\_para\_avg[i] = \rho \cdot sq\_para\_avg[i] + (1 - \rho) \cdot \text{change}^2
   $$

3. **New Solution**: 
   The new solution is calculated by subtracting the change from the current solution.

4. **Storing Solutions**:
   - The new solution is converted back to an array and appended to the solutions list.
   - The value of the objective function at the new solution is calculated and printed.

```python
from math import sqrt
from numpy import asarray
from numpy import arange
from numpy.random import rand
from numpy.random import seed
from numpy import meshgrid
from matplotlib import pyplot
from mpl_toolkits.mplot3d import Axes3D

# Objective function
def objective(x, y):
 return x**2.0 + y**2.0

# Derivative of objective function
def derivative(x, y):
 return asarray([x * 2.0, y * 2.0])

# Gradient descent algorithm with Adadelta
def adadelta(objective, derivative, bounds, n_iter, rho, ep=1e-3):
 solutions = list()
 solution = bounds[:, 0] + rand(len(bounds)) * (bounds[:, 1] - bounds[:, 0])
 sq_grad_avg = [0.0 for _ in range(bounds.shape[0])]
 sq_para_avg = [0.0 for _ in range(bounds.shape[0])]
 for it in range(n_iter):
  gradient = derivative(solution[0], solution[1])
  for i in range(gradient.shape[0]):
   sg = gradient[i]**2.0
   sq_grad_avg[i] = (sq_grad_avg[i] * rho) + (sg * (1.0 - rho))
  new_solution = list()
  for i in range(solution.shape[0]):
   alpha = (ep + sqrt(sq_para_avg[i])) / (ep + sqrt(sq_grad_avg[i]))
   change = alpha * gradient[i]
   sq_para_avg[i] = (sq_para_avg[i] * rho) + (change**2.0 * (1.0 - rho))
   value = solution[i] - change
   new_solution.append(value)
  solution = asarray(new_solution)
  solutions.append(solution)
  solution_eval = objective(solution[0], solution[1])
  print('>%d f(%s) = %.5f' % (it, solution, solution_eval))
 return solutions

# Seed the pseudo random number generator
seed(1)
# Define range for input
bounds = asarray([[-1.0, 1.0], [-1.0, 1.0]])
# Define the total iterations
n_iter = 120
# Rho for Adadelta
rho = 0.99
# Perform the gradient descent search with Adadelta
solutions = adadelta(objective, derivative, bounds, n_iter, rho)
# Sample input range uniformly at 0.1 increments
xaxis = arange(bounds[0,0], bounds[0,1], 0.1)
yaxis = arange(bounds[1,0], bounds[1,1], 0.1)
# Create a mesh from the axis
x, y = meshgrid(xaxis, yaxis)
# Compute targets
results = objective(x, y)
# Create a filled contour plot with 50 levels and jet color scheme
pyplot.contourf(x, y, results, levels=50, cmap='jet')
# Plot the sample as black circles
solutions = asarray(solutions)
pyplot.plot(solutions[:, 0], solutions[:, 1], '.-', color='w')
# Show the plot
pyplot.show()
```
